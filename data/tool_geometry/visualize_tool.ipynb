```python
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot



# Show the plot to the screen
camera_positions = [(0,0), (90,0), (90,90)]

# Create a new plot
figure = pyplot.figure(figsize=(30,30))

# Load the STL files and add the vectors to the plot
# your_mesh = mesh.Mesh.from_file('concrete_extruder_v1_10mm_0deg.stl')
#earth_extruder_v1_10mm_15deg.stl
your_mesh = mesh.Mesh.from_file('concrete_extruder_long_nozzle1.stl')

# Auto scale to the mesh size
scale = your_mesh.points.flatten()

# tool_centerpoint = (-5.04268e-10,-0.418261,0.14338)
# tool_orientation_x = (-1.57075e-09,-0.415647,0.153136)
# # tool_orientation_x = (-1.58844e-09,-0.420875,0.133625)
# tool_orientation_y = (1, 0.0, 0.0,)

# tool_centerpoint = (0,-0.154355,0.361564)
# tool_orientation_x = (0,-0.149169,0.367692)

tool_centerpoint = (0,-0.2347,0.4291)
tool_orientation_x = (0,-0.2295,0.4352)
tool_orientation_y = (1, 0.0, 0.0,)
# vec = Vector.from_start_end(tool_centerpoint, tool_orientation_x)
# ee_frame = Frame(tool_centerpoint, tool_orientation_y, vec)


vec_x = [tool_centerpoint[0], tool_orientation_x[0]]
vec_y = [tool_centerpoint[1], tool_orientation_x[1]]
vec_z = [tool_centerpoint[2], tool_orientation_x[2]]
#     vec_x = [6.09775e-10, 6.09777e-10]*100
# vec_y = [-0.418261, -0.419581]*100
# vec_z = [0.121892, 0.126818]*100
#add a vector

# axes.quiver(0,0,0,0,-154.312,361.564,normalize=False, color='red')

for i, (angle, axis) in enumerate(camera_positions):
    axes = figure.add_subplot(3, 3, i+1, projection='3d')

    axes.add_collection3d(mplot3d.art3d.Poly3DCollection(your_mesh.vectors, alpha=0.15))
    axes.set_xlabel('x')
    axes.set_ylabel('y')
    axes.set_zlabel('z')
    axes.plot(vec_x, vec_y, vec_z, color='red')
    axes.axis('equal')
    # axes.set(xlim=(-0.2, 0.2), ylim=(-1, 0.6), zlim=(0,0.4))

    axes.view_init(azim=angle, elev=axis)
    
# Show all the plots
pyplot.show()
```
